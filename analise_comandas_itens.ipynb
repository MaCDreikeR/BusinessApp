## 4. Implementar Valida√ß√£o de Tipos

Vamos criar fun√ß√µes de valida√ß√£o para os tipos existentes e definir regras para o novo tipo "pagamento".

```python
# Definir tipos v√°lidos e suas regras
TIPOS_VALIDOS = {
    'produto': {
        'nome': 'Produto',
        'descricao': 'Item f√≠sico vendido',
        'requer_item_id': True,
        'requer_quantidade': True,
        'permite_estoque': True,
        'icone': 'üì¶',
        'cor': '#7C3AED'
    },
    'servico': {
        'nome': 'Servi√ßo',
        'descricao': 'Servi√ßo prestado',
        'requer_item_id': True,
        'requer_quantidade': True,
        'permite_estoque': False,
        'icone': 'üîß',
        'cor': '#10B981'
    },
    'pacote': {
        'nome': 'Pacote',
        'descricao': 'Conjunto de produtos/servi√ßos',
        'requer_item_id': True,
        'requer_quantidade': True,
        'permite_estoque': False,
        'icone': 'üì¶',
        'cor': '#F59E0B'
    },
    'pagamento': {
        'nome': 'Pagamento',
        'descricao': 'Valor pago diretamente',
        'requer_item_id': False,
        'requer_quantidade': False,
        'permite_estoque': False,
        'icone': 'üí≥',
        'cor': '#EF4444'
    }
}

def validar_item_comanda(item_data):
    """
    Valida um item de comanda com base no tipo
    """
    tipo = item_data.get('tipo')
    
    if tipo not in TIPOS_VALIDOS:
        return False, f"Tipo '{tipo}' n√£o √© v√°lido. Tipos permitidos: {list(TIPOS_VALIDOS.keys())}"
    
    regras = TIPOS_VALIDOS[tipo]
    erros = []
    
    # Validar item_id
    if regras['requer_item_id'] and not item_data.get('item_id'):
        erros.append(f"Tipo '{tipo}' requer item_id")
    
    # Validar quantidade
    if regras['requer_quantidade']:
        quantidade = item_data.get('quantidade', 0)
        if not quantidade or quantidade <= 0:
            erros.append(f"Tipo '{tipo}' requer quantidade v√°lida")
    
    # Validar pre√ßo
    preco = item_data.get('preco_unitario', 0)
    if not preco or preco <= 0:
        erros.append("Pre√ßo unit√°rio deve ser maior que zero")
    
    # Validar nome
    if not item_data.get('nome', '').strip():
        erros.append("Nome √© obrigat√≥rio")
    
    if erros:
        return False, "; ".join(erros)
    
    return True, "Item v√°lido"

# Teste de valida√ß√£o
print("üß™ TESTE DE VALIDA√á√ÉO:\n")

# Teste item v√°lido existente
item_produto = {
    'tipo': 'produto',
    'item_id': 'uuid-123',
    'quantidade': 1,
    'preco_unitario': 100.0,
    'nome': 'Produto Teste'
}

valido, msg = validar_item_comanda(item_produto)
print(f"‚úÖ Produto: {valido} - {msg}")

# Teste item pagamento (novo tipo)
item_pagamento = {
    'tipo': 'pagamento',
    'item_id': None,  # N√£o requer item_id
    'quantidade': 1,
    'preco_unitario': 200.0,
    'nome': 'Pagamento'
}

valido, msg = validar_item_comanda(item_pagamento)
print(f"üí≥ Pagamento: {valido} - {msg}")

# Teste item inv√°lido
item_invalido = {
    'tipo': 'produto',
    'item_id': None,  # Produto requer item_id
    'quantidade': 0,   # Quantidade inv√°lida
    'preco_unitario': -10,  # Pre√ßo inv√°lido
    'nome': ''  # Nome vazio
}

valido, msg = validar_item_comanda(item_invalido)
print(f"‚ùå Item inv√°lido: {valido} - {msg}")
```

## 5. Propor Nova Estrutura com Tipo "Pagamento"

Vamos definir a nova estrutura da tabela que inclui o tipo "pagamento" e resolve os problemas identificados.

```python
# Estrutura nova proposta
nova_estrutura = {
    'campos_obrigatorios': [
        'id',  # UUID primary key
        'comanda_id',  # UUID foreign key
        'tipo',  # ENUM: produto, servico, pacote, pagamento
        'nome',  # TEXT - nome do item
        'preco_unitario',  # DECIMAL - pre√ßo por unidade
        'quantidade',  # INTEGER - quantidade (padr√£o 1)
        'preco_total',  # DECIMAL - calculado: preco_unitario * quantidade
        'estabelecimento_id',  # UUID foreign key
        'created_at',  # TIMESTAMP
        'updated_at'   # TIMESTAMP
    ],
    'campos_opcionais': [
        'item_id',  # UUID - NULL para pagamentos
        'observacoes'  # TEXT - observa√ß√µes espec√≠ficas do item
    ],
    'constraints': [
        "CHECK (tipo IN ('produto', 'servico', 'pacote', 'pagamento'))",
        "CHECK (preco_unitario > 0)",
        "CHECK (quantidade > 0)",
        "CHECK (preco_total = preco_unitario * quantidade)",
        "CHECK ((tipo = 'pagamento' AND item_id IS NULL) OR (tipo != 'pagamento' AND item_id IS NOT NULL))"
    ]
}

print("üèóÔ∏è  NOVA ESTRUTURA PROPOSTA:\n")

print("üìã **Campos Obrigat√≥rios:**")
for campo in nova_estrutura['campos_obrigatorios']:
    print(f"  ‚úÖ {campo}")

print("\nüìã **Campos Opcionais:**")
for campo in nova_estrutura['campos_opcionais']:
    print(f"  üî∏ {campo}")

print("\nüîí **Constraints/Valida√ß√µes:**")
for i, constraint in enumerate(nova_estrutura['constraints'], 1):
    print(f"  {i}. {constraint}")

print("\n" + "="*60)

# Exemplo de registros na nova estrutura
print("\nüìù **EXEMPLO DE REGISTROS NA NOVA ESTRUTURA:**\n")

exemplos_novos = [
    {
        'id': 'uuid-1',
        'comanda_id': 'comanda-uuid-1',
        'tipo': 'produto',
        'nome': 'Notebook Gamer',
        'preco_unitario': 2500.00,
        'quantidade': 1,
        'preco_total': 2500.00,
        'item_id': 'produto-uuid-123',
        'estabelecimento_id': 'estab-uuid-1',
        'observacoes': None
    },
    {
        'id': 'uuid-2',
        'comanda_id': 'comanda-uuid-1',
        'tipo': 'servico',
        'nome': 'Formata√ß√£o Windows',
        'preco_unitario': 150.00,
        'quantidade': 1,
        'preco_total': 150.00,
        'item_id': 'servico-uuid-456',
        'estabelecimento_id': 'estab-uuid-1',
        'observacoes': 'Incluir antiv√≠rus'
    },
    {
        'id': 'uuid-3',
        'comanda_id': 'comanda-uuid-1',
        'tipo': 'pagamento',
        'nome': 'Pagamento',
        'preco_unitario': 500.00,
        'quantidade': 1,
        'preco_total': 500.00,
        'item_id': None,  # NULL para pagamentos!
        'estabelecimento_id': 'estab-uuid-1',
        'observacoes': 'Entrada do servi√ßo'
    }
]

df_novo = pd.DataFrame(exemplos_novos)
print("üìä Amostra da nova estrutura:")
print(df_novo[['tipo', 'nome', 'preco_unitario', 'item_id', 'observacoes']].to_string(index=False))
```

## 7. Simular Inser√ß√£o de Registros de Pagamento

Vamos demonstrar como registros de pagamento seriam inseridos e como isso melhoraria o sistema.

```python
# Simular uma comanda completa com pagamentos
def simular_comanda_com_pagamento():
    comanda_exemplo = [
        # Produtos/Servi√ßos tradicionais
        {
            'id': 'item-1',
            'comanda_id': 'comanda-123',
            'tipo': 'produto',
            'nome': 'Notebook Dell',
            'preco_unitario': 3500.00,
            'quantidade': 1,
            'preco_total': 3500.00,
            'item_id': 'produto-uuid-dell',
            'estabelecimento_id': 'estab-1',
            'observacoes': None
        },
        {
            'id': 'item-2',
            'comanda_id': 'comanda-123',
            'tipo': 'servico',
            'nome': 'Instala√ß√£o de Software',
            'preco_unitario': 200.00,
            'quantidade': 1,
            'preco_total': 200.00,
            'item_id': 'servico-uuid-install',
            'estabelecimento_id': 'estab-1',
            'observacoes': 'Incluir Office e antiv√≠rus'
        },
        # NOVIDADE: Pagamentos diretos
        {
            'id': 'item-3',
            'comanda_id': 'comanda-123',
            'tipo': 'pagamento',
            'nome': 'Entrada (50%)',
            'preco_unitario': 1850.00,
            'quantidade': 1,
            'preco_total': 1850.00,
            'item_id': None,  # NULL para pagamentos!
            'estabelecimento_id': 'estab-1',
            'observacoes': 'Entrada paga em dinheiro'
        },
        {
            'id': 'item-4',
            'comanda_id': 'comanda-123',
            'tipo': 'pagamento',
            'nome': 'Restante (50%)',
            'preco_unitario': 1850.00,
            'quantidade': 1,
            'preco_total': 1850.00,
            'item_id': None,  # NULL para pagamentos!
            'estabelecimento_id': 'estab-1',
            'observacoes': 'A pagar na entrega'
        }
    ]
    
    return pd.DataFrame(comanda_exemplo)

# Executar simula√ß√£o
df_simulacao = simular_comanda_com_pagamento()

print("üí∞ SIMULA√á√ÉO: COMANDA COM PAGAMENTOS\n")
print("üìã Itens da comanda:")
print(df_simulacao[['tipo', 'nome', 'preco_unitario', 'item_id', 'observacoes']].to_string(index=False))

# An√°lise financeira da comanda
print(f"\nüìä AN√ÅLISE FINANCEIRA:")

total_produtos_servicos = df_simulacao[df_simulacao['tipo'].isin(['produto', 'servico'])]['preco_total'].sum()
total_pagamentos = df_simulacao[df_simulacao['tipo'] == 'pagamento']['preco_total'].sum()
saldo_pendente = total_produtos_servicos - total_pagamentos

print(f"  üíº Total de produtos/servi√ßos: R$ {total_produtos_servicos:,.2f}")
print(f"  üí≥ Total de pagamentos: R$ {total_pagamentos:,.2f}")
print(f"  üìâ Saldo pendente: R$ {saldo_pendente:,.2f}")

# Visualiza√ß√£o da comanda
plt.figure(figsize=(12, 6))

# Gr√°fico 1: Composi√ß√£o por tipo
plt.subplot(1, 2, 1)
composicao = df_simulacao.groupby('tipo')['preco_total'].sum()
cores = ['#EF4444', '#7C3AED', '#10B981']  # pagamento, produto, servico
composicao.plot(kind='pie', autopct='%1.1f%%', colors=cores, startangle=90)
plt.title('üí∞ Composi√ß√£o da Comanda\\n(Com Pagamentos)', fontweight='bold')
plt.ylabel('')

# Gr√°fico 2: Fluxo de caixa
plt.subplot(1, 2, 2)
categorias = ['Produtos/Servi√ßos', 'Pagamentos Recebidos', 'Saldo Pendente']
valores = [total_produtos_servicos, total_pagamentos, saldo_pendente]
cores_fluxo = ['#3B82F6', '#10B981', '#F59E0B']
bars = plt.bar(categorias, valores, color=cores_fluxo)

# Adicionar valores nas barras
for bar, valor in zip(bars, valores):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 50,
             f'R$ {valor:,.0f}', ha='center', va='bottom', fontweight='bold')

plt.title('üìä Fluxo de Caixa da Comanda', fontweight='bold')
plt.ylabel('Valor (R$)')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

print(f"\nüéØ **BENEF√çCIOS DOS PAGAMENTOS:**")
print(f"  ‚úÖ Controle de entradas parciais")
print(f"  ‚úÖ Hist√≥rico completo de recebimentos")
print(f"  ‚úÖ C√°lculo autom√°tico de saldos")
print(f"  ‚úÖ Relat√≥rios financeiros precisos")
print(f"  ‚úÖ Melhor gest√£o de fluxo de caixa")
```
